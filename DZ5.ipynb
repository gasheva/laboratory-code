```python
#Задание 1
class Car:    
    def __init__(self, mass, engine_volume, power, price):
        self.__mass=mass
        self.__engine_volume=engine_volume
        self.__power=power
        self.__price=price
    #свойства для массы
    @property
    def mass(self):
        return self.__mass
    @mass.setter
    def mass(self, m):
        self.__mass = m
    #свойства для объема двигателя
    @property
    def engine_volume(self):
        return self.__engine_volume
    @engine_volume.setter
    def engine_volume(self, v):
        self.__engine_volume = v
    #свойства для мощности
    @property
    def power(self):
        return self.__power
    @power.setter
    def power(self, p):
        self.__power = p
    #свойства для цены
    @property
    def price(self):
        return self.__price
    @price.setter
    def price(self, p):
        self.__price = p
    #печатаем "бип"
    def printBip(self):
        print("БИП!")
    
class Truck(Car):
    def __init__(self, trailer, mass, engine_volume, power, price):
        super().__init__(mass, engine_volume, power, price)
        self.__trailer=trailer
       
    @property
    def trailer(self):
        return self.__trailer
    @trailer.setter
    def trailer(self, t: bool):
        if type(t)!="Boolean":
            raise TypeError("Ожидалось boolean значение")
        else:
            self.__trailer=t
        
    def printBip(self):
        print("БИП-БИП!")
    def unloadTruck(self):
        print("Грузовик разгружен")

print("---Авто---")
car = Car(1000, 4.5, 89, 550)
print("Масса = "+str(car.mass))
print("Объем = "+str(car.engine_volume))
print("Мощность = "+str(car.power))
print("Цена = "+str(car.price))
car.printBip()

print("\n---Грузовик---")
truck = Truck(True, 5000, 9.5, 160, 1500)
truck.mass = 3000
print("Масса = "+str(truck.mass))
print("Объем = "+str(truck.engine_volume))
print("Мощность = "+str(truck.power))
print("Цена = "+str(truck.price))
has_truck_print = lambda x: "да" if x==True else "нет"
print("Есть прицеп = "+ has_truck_print(truck.trailer))
truck.printBip()
truck.unloadTruck()

truck.trailer=4
```

```python
#Задание 2
class AuthorsIterator:
    def __init__(self, *authors: list):
        self.__authors = list()
        self.__counter = 0
        if authors:
            self.__authors.append(authors)
    @property
    def authors(self):
        return self.__authors
    @authors.setter
    def authors(self, arr: list):
        self.__authors = arr
    
    def add_author(self, author):
        self.__authors.append(author)
        
    #методы для итерации
    def __iter__(self):
        return iter(self.__authors)

    def __next__(self):
        if self.__counter < len(self.__authors):
            self.__counter += 1
            return 1
        else:
            raise StopIteration

            
class AuthorsGenerator:
    def __init__(self, *authors: list):
        self.__authors = list()
        self.__counter = 0
        if authors:
            self.__authors.append(authors)
    @property
    def authors(self):
        return self.__authors
    @authors.setter
    def authors(self, arr: list):
        self.__authors = arr
    
    def add_author(self, author):
        self.__authors.append(author)
        
    #методы для итерации
    def __iter__(self):
        for elem in self.__authors:
            yield elem

    def __next__(self):
        if self.__counter < len(self.__authors):
            self.__counter += 1
            return 1
        else:
            raise StopIteration
    
    
#класс, хранящий инфу о писателе
class Author:
    def __init__(self, name, surname, country, *otchestvo):
        self.__name=name
        self.__surname=surname
        self.__country=country
        if otchestvo:
            self.__otchestvo=otchestvo[0]
        else:
            self.__otchestvo=None
       
    #свойства...
    #метод для печати инфы о писателе
    def print_author(self):
        print(self.__name+" "+self.__surname + ((" "+self.__otchestvo) if self.__otchestvo else "") +" from "+ self.__country)

milne = Author("Alan", "Milne", "England")
milne.print_author()
pushkin = Author("Aleksandr", "Pushkin", "Russia", "Sergeevich")

auth_it = AuthorsIterator()
auth_it.add_author(milne)
auth_it.add_author(pushkin)

#печатаем авторов из списка
print("\n---Выводим список с итератором---")
for a in auth_it:
    a.print_author()
    
auth_gen = AuthorsGenerator()
auth_gen.add_author(milne)
auth_gen.add_author(pushkin)

#печатаем авторов из списка
print("\n---Выводим список с генератором---")
for a in auth_gen:
    a.print_author()
    
auth_gen.add_author(milne)
#печатаем авторов из списка
print("\n---Выводим список с генератором после добавления еще одного писателя---")
for a in auth_gen:
    a.print_author()
```
